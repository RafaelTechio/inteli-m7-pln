# Adicionando 2 Features para cada dataset filtrado


## Cozinha

**asking_for_recipe** -> Booleano que indica se a pergunta é o pedido de uma receita

**asking_for_doubt** -> Booleano que indica se a pergunta/pedido está relacionado com uma dúvida

```python
df_cooking['asking_for_recipe'] = df_cooking['utt'].str.contains(r'\b(almoço|janta|café da manhã|brunch|receita|forma)\b')
df_cooking['asking_for_doubt'] = df_cooking['utt'].str.contains(r'\b(usar|acrescentar|tirar|pode)\b')

df_cooking
```

## Musica

**asking_for_album** -> Booleano que indica se a pergunta/pedido está relacionado com o album da música

**asking_for_autor** -> Booleano que indica se a pergunta/pedido está relacionado com quem canta a música, seja uma banda, grupo etc

```python
df_music['asking_for_album'] = df_music['utt'].str.contains(r'\b(álbum|disco)\b')
df_music['asking_for_autor'] = df_music['utt'].str.contains(r'\b(banda|grupo|dupla|autor|trio|cantor|cantora|artista)\b')

df_music
```

## Notícias

**asking_for_channel** -> Booleano que indica se a pergunta/pedido está relacionado com algum canal específico

**category** -> Categoria da notícia pedida (soccer ou tech)

```python
df_news['asking_for_channel'] = df_news['utt'].str.contains(r'\b(canal|notícias)\b')

about_soccer = df_news['utt'].str.contains(r'\b(futebol|bola|flamengo|corinthians)\b')
about_tech = df_news['utt'].str.contains(r'\b(tech|tecnologia)\b')

df_news.loc[about_soccer, 'category'] = 'soccer'
df_news.loc[about_tech, 'category'] = 'tech'

df_news
```
